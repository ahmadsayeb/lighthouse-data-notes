```python
imgmat = np.array(myImage)
lx,ly,lz = imgmat.shape
```

```python
X,Y = np.ogrid[0:lx,0:ly]
```

```python
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4
```

```python
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0
```

```python
plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```
